第一題

```python
import time
import matplotlib.pyplot as plt

def fibonacci_recursive(n):
    """使用頂部向下（純遞迴）方法計算第n個斐波那契數"""
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

def fibonacci_dynamic(n):
    """使用底部向上（動態規劃）方法計算第n個斐波那契數"""
    fib = [0, 1]
    for i in range(2, n + 1):
        fib.append(fib[i - 1] + fib[i - 2])
    return fib[n]

n_values = list(range(1, 101))
recursive_times = []
dynamic_times = []

for n in n_values:
    start_time = time.time()
    fibonacci_recursive(n)
    end_time = time.time()
    recursive_times.append(end_time - start_time)

    start_time = time.time()
    fibonacci_dynamic(n)
    end_time = time.time()
    dynamic_times.append(end_time - start_time)

plt.plot(n_values, recursive_times, label='Recursive')
plt.plot(n_values, dynamic_times, label='Dynamic')
plt.xlabel('n')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Fibonacci Sequence')
plt.legend()
plt.show()
```

第二題

```python
def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n + 1):
        fib.append(fib[i - 1] + fib[i - 2])
    return fib

fib_series = fibonacci(50)
times_computed = []
for n in range(5, 51):
    times_computed.append(fib_series[n - 4])

```

```python
import matplotlib.pyplot as plt

x = list(range(5, 51))
y = times_computed

plt.plot(x, y)
plt.xlabel('n')
plt.ylabel('Number of times F(4) is computed')
plt.title('Number of times F(4) is computed for F(5) to F(50)')
plt.show()
```
